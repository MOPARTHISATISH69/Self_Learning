Confusion Matrix: -

Source :: 

https://intellipaat.com/community/5048/how-to-write-a-confusion-matrix-in-python

https://stackoverflow.com/questions/2148543/how-to-write-a-confusion-matrix-in-python

```python
# Changing the float datatype to int. 
actual_y = np.array(df['y'])               
actual_y = actual_y.astype('int64')
pred_y = np.array(df['proba'])
pred_y = pred_y.astype('int64')

def compute_confusion_matrix(true, pred):
    # finding number of classes
    K = len(np.unique(true)) 
    # creating a resultant zeros matrix
    result = np.zeros((K, K))

    # putting values in each of the indices
    for i in range(len(true)):
        result[true[i]][pred[i]] += 1

    return result

confusion_matrix = compute_confusion_matrix(actual_y, pred_y)
print(confusion_matrix)
```

```python
def calculate_f1(confusion_matrix):
    # fetching values from confusion matrix
    true_positive,false_positive,false_negative = confusion_matrix[1,1],confusion_matrix[1,0],confusion_matrix[0,1]
    
    # calculating precision
    precision = true_positive / (true_positive + false_positive)
    # calculating recall
    recall = true_positive / (true_positive + false_negative)
    # calculating f1-score
    f1 = 2 * ((precision * recall) / (precision + recall))
    
    return f1


f1 = calculate_f1(confusion_matrix)
print(f1)
```

**1.3 AUC**

Source :

https://machinelearningmastery.com/roc-curves-and-precision-recall-curves-for-classification-in-python/

```python
df = pd.read_csv('5_a.csv')
```

```python
def auc_calc(df):
    # first creating a list of all the unique y_pred's
    Unique_proba = list(df['proba'].unique())
    # sorting them in descending order
    Unique_proba.sort(reverse = True)
    
    # initializing two empty lists for storing 'tpr' and 'fpr'
    tpr=[]
    fpr=[]


    for thresh in Unique_proba:                                               # iterating through each of the sorted array values
        threshold_values_probability = []                                     # initializing an empty list to store thresholded values
        for probability in df['proba']:                                       # comparing this threshold value with each of the value in y_pred
            if probability > thresh:                                         
                threshold_values_probability.append(1)
            else:
                threshold_values_probability.append(0)
            
        df['thresholding'] = threshold_values_probability                                       # adding a new column
    
       
        TN = (((df['y'])==0)& ((df['thresholding'])==0)).sum()
        TP = (((df['y'])==1) & ((df['thresholding'])==1)).sum()                   # performing coparisions to determine tpr and fpr
        FP = (((df['y'])==0) & ((df['thresholding'])==1)).sum()
        FN = (((df['y'])==1) & ((df['thresholding'])==0)).sum()
    
        tpr.append(TP/(TP+FN))                                                # appending results to the tpr and fpr list
        fpr.append(FP/(FP+TN))
    
    # calculating 'auc'
    a = sorted(tpr)
    b = sorted(fpr)
    auc = np.trapz(a,b)
    
    return auc


auc = auc_calc(df)
print(auc)
```

```python
## 1.4 Accuracy Score:

def calculate_acc(confusion_matrix):
    # using confusion matrix fetching values
    true_positive = confusion_matrix[1,1]
    true_negative = confusion_matrix[0,0]
    false_positive = confusion_matrix[1,0]
    false_negative = confusion_matrix[0,1]
    
    # computing accuracy score
    acc = (true_negative + true_positive) / (true_positive + true_negative + false_positive + false_negative)
    
    return acc


accuracy_score = calculate_acc(confusion_matrix)
print(accuracy_score)
```

<font color='red'><b>C.</b></font> Compute the best threshold (similarly to ROC curve computation) of probability which gives lowest values of metric <b>A</b> for the given data <strong>5_c.csv</strong>
<br>

you will be predicting label of a data points like this: $y^{pred}= \text{[0 if y_score < threshold  else 1]}$

$ A = 500 \times \text{number of false negative} + 100 \times \text{numebr of false positive}$

<pre>
   <b>Note 1:</b> in this data you can see number of negative points > number of positive points
   <b>Note 2:</b> use pandas or numpy to read the data from <b>5_c.csv</b>
</pre>

```python
 # write your code

 df = pd.read_csv('5_c.csv')
```

```python
def find_threshold(df):
    unique_df = list(df['prob'].unique())                                 ## finding unique values form 'prob' column, storing them to a list
    unique_df.sort(reverse = True)                                        ## Sorting in descending order  
    
    lst = []                                                              ## initializing an empty list to store all the operations
    for thresh in unique_df:                                              ## for each threshold value of sorted df
        new_y_pred = []
        for val in df['prob']:                                            ## comparing it with each of the value of 'prob' column
            if val > thresh:
                new_y_pred.append(1)
            else:
                new_y_pred.append(0)
    
        # adding it to the df
        df['thresholding'] = new_y_pred

        # performing comparision
        FP = (((df['y'])==0) & ((df['thresholding'])==1)).sum()
        FN = (((df['y'])==1) & ((df['thresholding'])==0)).sum()
    
       
        Conditions_apply = (500 * FN) + (100 * FP)                                   # applying conditions
        lst.append(Conditions_apply)                                                 # appending them to a list
    
    
    lowest = np.argmin(lst)                                      # finding the minimum value's index from the list
    low_val = min(lst)                                           # finding the minimum value
    best = unique_df[lowest]                                     # finding the best threshold value
    
    return low_val, best



low_value,best = find_threshold(df)
print(low_value)
print(best)
```

```python
# reading csv for Task-4
df = pd.read_csv('5_d.csv')
df.head()
```

```python
## 4.1 Mean Square Error

def MSE(df):
    mean_square_error = 0
    for idx, row in df.iterrows():                            # taking each row of the df
        act = row['y']                                        # actual value
        pred = row['pred']                                    # predicted value
        mean_square_error += (act - pred) ** 2                              # fining squared error
    
    # calculating final results
    final = mean_square_error / df.shape[0]
    return final


mse = MSE(df)
print(mse)
```

```python
## 4.2 Mean Absolute Percentage Error

def MAPE(df):
    absolute_error = 0
    actual_true = 0
    for idx, row in df.iterrows():                                 # taking each row from df iteratively
        actual = row['y']                                             # actual value
        pred = row['pred']                                         # predicted value
        absolute_error += abs(actual - pred)                                       # finding absolute error
        actual_true +=  actual                                           # finding sum of actual values
    
    # calculating final results
    final = (absolute_error / actual_true)
    return final



mape = MAPE(df)
print(mape)
```

```python
## 4.3 R^2

def cod(df):
    residuals_sum = 0
    sum_of_squares = 0
    mean_value = np.mean(df['y'])                                    # calculating mean of all the y's
    for idx, row in df.iterrows():                                   # taking each row from df iteratively
        act = row['y']                                               # actual value
        pred = row['pred']                                           # predicted value
        residuals_sum += (act - pred) ** 2                           # calculating sum of residuals
        sum_of_squares += (act - mean_value) ** 2                    # calculating sum of squares
        
    # calculating final results
    final = 1 - (residuals_sum / sum_of_squares)
    return final


R_square = cod(df)
print(R_square)
```
